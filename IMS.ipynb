```python
print("****************** INVENTORY MANAGEMENT SYSTEM ******************")
while(int(input("SELECT:\n 1.SHOP\n 0.EXIT\n"))!=0):
    print("YOU ARE:\n 1. ADMIN\n 2. CUSTOMER\n")
    s= int(input())
    print("\n")
    if(s==1):
        print("1. ADD NEW PRODUCTS\n 2. DELETE A PRODUCT\n 3. DISPLAY SALES LIST\n")
        c=int(input("Enter your choice:"))
        print("\n")
        
        if(c==1):
            print("***ADD NEW PRODUCT***\n")
            A_id = input("Enter the product id: ")
            A_name = input("Enter the product name: ")
            A_cost =input("Enter the cost: ")
            A_quantity = input("Enter the product quantity: ")
            A_brand = input("Enter the brand: ")
            A_expiry = input("Enter the date of expiry: ")
            record[A_id] = { "Name" : A_name , "Cost" : A_cost , "Quantity" : A_quantity , "Brand" : A_brand  ,  "Expiry" : A_expiry  }
            print("Product has been added successfully!\n")
        
        if(c==2):
            A_id = input("Enter the product id to delete ")
            record.pop(A_id)
            print("Product has been deleted successfully....\n")
            
        if(c==3):
            ("****** SALES LIST  ******\n")
            fdr = open("sales.json" , 'r')
            txt1 = fdr.read()
            fdr.close()
            print(txt1)
    
    if(s==2):
        cname = input("Enter the Customer name : ")
        order_no = int (input("Enter the product id : "))
        o_quantity = int (input("Enter the quantity of product : "))
        
        record[str(order_no)]["Quantity"] = record[str(order_no)]["Quantity"] - o_quantity
        amount= record[str(order_no)]["Cost"]*o_quantity
        
        print("\n*************************************************\n")
        print("*********BILL********\n")
        print("Customer Name : " + " "+ cname)
        print("Product Name : "+ " "+ record[str(order_no)]["Name"])
        print("Product Id : " +" "+  str(order_no))
        print("Quantity : " +" "+ str(o_quantity))
        print("Billing amount : " +""+str(amount))
        print("\n")
        print("********** THANX FOR ORDERING! **********")
        print("\n")
        
        
        js = json.dumps(record)
        fd = open("record.json",'w')
        fd.write(js)
        fd.close()
        
        
        sales={"Customer name":cname, "Prod_id":order_no,"Prod_name":record[str(order_no)]["Name"], "quantity":o_quantity, "amount":amount,"Time":time.ctime()}
        fs = open("sales.json",'a')
        sd=json.dumps(sales)
        fs.write(sd)
        fs.close()
       
```
